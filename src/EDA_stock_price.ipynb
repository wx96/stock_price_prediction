```python
import pandas as pd

import plotly.express as px
import plotly.graph_objs as go
import matplotlib.pyplot as plt
```

```python
def plot_trend(data, x, y, legend):
    fig = px.line(data, x=x, y=y,color=legend)
    fig.show()
```

```python
plot_trend(stock_nyse,'Date','Value','Ticker')
```

Stock Price Exxon

```python
stock_exxon = pd.read_csv('../sample_dataset/nyse-energy-company-stock-closing-prices-2020.csv', parse_dates=True)
#stock_exxon.set_index('Date', inplace=True)
stock_exxon['Date'] = pd.to_datetime(stock_exxon['Date'])
stock_exxon = stock_exxon.loc[stock_exxon['Ticker']=='XOM']
print(stock_exxon.shape)
display(stock_exxon.head(5))
#stock_exxon['Value'].plot(figsize=(12,6))
plt.plot(stock_exxon['Date'],stock_exxon['Value'])
plt.show()
```

### Stock index

1. <b>Dow Jones Utilities Average</b>, DJUA is a stock index from S&P Dow Jones Indices that tracks the performance of 15 prominent utility companies traded in the United States.
2. <b>Dow Jones Transportation Average</b>, DJTA is a price-weighted average of 20 transportation stocks traded in the United States.
3. <b>Dow Jones Industrial Average</b>, DJIA is a stock market index that measures the stock performance of 30 large companies listed on stock exchanges in the United States.
4. <b>Dow Jones Composite Average</b> is a stock index from Dow Jones Indexes that tracks 65 prominent companies. The average's components include every stock from the Dow Jones Industrial Average (30 components), the Dow Jones Transportation Average (20), and the Dow Jones Utility Average (15)
5. <b>S&P500 (Standard & Poor 500)</b> is a stock market index that measures the stock performance of 500 large companies listed on stock exchanges in the United States. It is one of the most commonly followed equity indices.

```python
stock_index = pd.read_csv('../sample_dataset/some-stock-market-indicators-&-index-2020.csv',parse_dates=True)
stock_index['Date'] = pd.to_datetime(stock_index['Date'])
print(stock_index.shape)
display(stock_index.head(5))
# display(stock_index.describe())
plot_trend(stock_index,'Date','Value','Description')
```

```python
snp500 = stock_index.loc[stock_index['Description']=='S&P 500']
exxon_snp500 = pd.merge(stock_exxon, snp500, left_on = ['Date'], right_on = ['Date'], how='left')

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Exxon', color=color)
ax1.plot(exxon_snp500['Date'], exxon_snp500['Value_x'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('S&P 500', color=color)  # we already handled the x-label with ax1
ax2.plot(exxon_snp500['Date'], exxon_snp500['Value_y'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
```

## Oil demand

```python
oil_demand = pd.read_csv('../sample_dataset/jodi-oil-demand-2020.csv', parse_dates=['Date'])
print(oil_demand.shape)
display(oil_demand.head(5))
display(oil_demand.describe())
```

```python
oil_demand_usa = oil_demand.loc[ oil_demand['Country'] == 'UNITED STATES'] 
#oil_demand_usa['Date'] = pd.to_datetime(oil_demand_usa['Date'])
oil_demand_exxon = pd.merge(stock_exxon, oil_demand_usa, left_on = ['Date'], right_on = ['Date'], how='left')

oil_demand_exxon['Value_y'].fillna(method='bfill')
print(oil_demand_exxon)
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Exxon', color=color)
ax1.plot(oil_demand_exxon['Date'], oil_demand_exxon['Value_x'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Oil Demand', color=color)  # we already handled the x-label with ax1
ax2.plot(oil_demand_exxon['Date'], oil_demand_exxon['Value_y'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
```
