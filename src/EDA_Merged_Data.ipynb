```python
combined_data = pd.read_csv('../data/data_2010_2021.csv', parse_dates=True)
print(combined_data.shape)
combined_data.head(2)
```

```python
print(combined_data.columns.tolist())
```

```python
combined_data_corr = pd.DataFrame(combined_data,columns=['stock_closing_usd','sentiment_global_index_x','sentiment_global_index_y',
                                                        'S&P 500','DOW JONES INDUSTRIAL AVERAGE','NYMEX CRUDE OIL FUTURES',
                                                        'JODI_demand_UNITED STATES'])

corrMatrix = combined_data_corr.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
combined_data_2017 = combined_data[combined_data['Date']>='2017-01-01']
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Exxon Stock Price', color=color)
ax1.plot(combined_data_2017['Date'], combined_data_2017['stock_closing_usd'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sentiment_global_index_x', color=color)  # we already handled the x-label with ax1
ax2.plot(combined_data_2017['Date'], combined_data_2017['sentiment_global_index_x'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
```
